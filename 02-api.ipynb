```python
# en esta lista almacenamos todos los datos que extrae y se guardan en datos_json

def llamar_api(inicio, final)
    datos_total = []
    for año in range(inicio, (final + 1)):
        
        # extraemos datos de la API
        datos = requests.get(f'https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date={año}-01-01T00:00&end_date={año}-12-31T23:59&time_trunc=month')
        datos_json = datos.json()

        # appendeamos en la lista
        datos_total.append(datos_json)

    return datos_total
```

```python
def extraccion_datos(lista_origen):
    # return de la función
    diccionario = {'tipo energia': [], 'tipo generacion': [], 'fecha': [], 'valor': []} 

    # listas donde appendear datos extraidos
    energia = []
    generacion = []
    fecha = []
    valor = []

    for dato in lista_origen:
        
        for i in dato['included']:
            
            # estos datos deben tener el mismo len que los del siguiente bucle para poder hacer un dataframe
            energia.append(i['attributes']['type'])
            generacion.append(i['type'])

            for e in i['attributes']['values']:
                fecha.append(e['datetime'])
                valor.append(e['value'])

       
    diccionario['tipo energia'] = energia
    diccionario['tipo generacion'] = generacion
    diccionario['fecha'] = fecha
    diccionario['valor'] = valor
    
    return diccionario

```

```python
diccionarios_pair = extraccion_datos(datos_energia)
```
